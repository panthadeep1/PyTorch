```python
## Author: @panthadeep_b
##Created on: 06.Oct.25; 20.57 IST
## Introduction to PyTorch

##Check for GPU
import torch
dev_info = torch.cuda.get_device_name(0);
dev = torch.device("cuda");
print("The device is:", dev_info);

#to check if your GPU driver and CUDA/ROCm is enabled and accessible by PyTorch
torch.cuda.is_available()

```

```python
x = torch.tensor([1.0, 2.0, 3.0]);
print('1D Tensor: \n', x);

y = torch.zeros((3, 3));
print('2D Tensor: \n', y);

r = torch.randn(3, 4, 5, device = dev, dtype = torch.float);
print('Random Tensor: \n', r);
```

```python
##How to create tensors
V_data = [1,2,3,4,5];
V = torch.tensor(V_data);
print("The tensor data is: ", V);
```

```python
##Create tensor of random data
x = torch.randn((2, 3, 4))
print(x)
```

```python
##Perform operations using torch

arr1 = torch.tensor([1.0, 2.0]);
arr2 = torch.tensor([3.0, 4.0]);

arr = arr1 + arr2;
print("The add matrix result is: ", arr);

e1 = arr1.view(2,1);
print("e1: ", e1);

e2 = arr2.view(1,2);
print("e2: ", e2);

mat_mul = torch.matmul(e1, e2);
print("The matrix multiplcation of e1, e2 is: ", mat_mul);
```
