```python
## Author: @panthadeep_b
## Created on: 02.Jan.26; 20.44 IST
## Gradient-Descent in Machine Learning

import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
#Set the input and output data
# set random seed for reproducibility
torch.manual_seed(42);

# set number of samples 
num_samples = 1000;

# create random features with 2-D
x = torch.randn(num_samples, 2);

# create random weights and bias for the linear regression model
true_weights = torch.tensor([1.3, -1]);
true_bias    = torch.tensor([-3.5]);

# Target variable
y = x @ true_weights.T + true_bias;

# Plot the dataset
fig, ax = plt.subplots(1, 2, sharey=True)
ax[0].scatter(x[:,0],y)
ax[1].scatter(x[:,1],y)

ax[0].set_xlabel('X1')
ax[0].set_ylabel('Y')
ax[1].set_xlabel('X2')
ax[1].set_ylabel('Y')
plt.show()
```
